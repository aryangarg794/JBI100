```python
import pandas as pd
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoConfig
```

```python
df_tweets_01 = pd.read_csv('../Data/FIFA World Cup 2022 Twitter Dataset/tweets1.csv', delimiter=';')
df_tweets_02 = pd.read_csv('../Data/FIFA World Cup 2022 Twitter Dataset/tweets2.csv', delimiter=';')
df_tweets = pd.concat([df_tweets_01, df_tweets_02])
```

```python
df_tweets.sort_values(by='Tweet Posted Time', ascending=False)
```

```python
df_tweets.head()
```

```python
MODEL = "cardiffnlp/twitter-roberta-base-sentiment-latest"

tokenizer = AutoTokenizer.from_pretrained(MODEL, model_max_len=512)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL)
```

```python
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer, config=config)
```

```python
tokenizer_kwargs = {'padding':True,'truncation':True,'max_length':512}
classifier("hello \
           ", **tokenizer_kwargs)[0]['label']
```

```python
tokenizer_kwargs = {'padding':True,'truncation':True,'max_length':510}
df_tweets['Sentiment'] = df_tweets['Tweet Content'].apply(lambda tweet: classifier(tweet, **tokenizer_kwargs)[0]['label'])
```

```python
df_tweets.to_csv('../Data/FIFA World Cup 2022 Twitter Dataset/tweets_sentiment.csv', index=False)
```
