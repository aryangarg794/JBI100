```python
import pandas as pd
```

```python
df_nameslist1 = pd.read_csv('//Data/FIFA World Cup 2022 Player Data/stats_combined.csv', delimiter=',', usecols=['player', 'position', 'team', 'age'])
df_nameslist2 = pd.read_csv('//Data/FIFA World Cup 2022 Player Data/stats_combined_keepers.csv', delimiter=',', usecols=['player', 'position', 'team', 'age'])
df_tweetdata = pd.read_csv('//Data/FIFA World Cup 2022 Twitter Dataset/tweets_sentiment.csv', delimiter=',', usecols=['Tweet Content', 'Sentiment'])
```

```python
full_list = [df_nameslist1, df_nameslist2]
df_nameslist = pd.concat(full_list)
df_nameslist
```

```python
df_tweetdata.head()
```

```python
result_dict = {'player' : [], 'position' : [], 'team' : [], 'age' : [], 'mentions_count' : [], 'sentiment_score' : [], 'positive_tweets' : [], 'neutral_tweets' : [], 'negative_tweets' : []}
iteration = 0
for name in df_nameslist['player']: # Calculates the number of mentions, sentiment_score, positive, neutral, and negative tweets per player
    name_parts = name.split()
    first_name = name_parts[0]
    last_name = ' '.join(name_parts[1:]) if len(name_parts) > 1 else name_parts[0]
    last_name = ' ' + last_name + ' '

    name_df = df_tweetdata[(df_tweetdata['Tweet Content'].str.lower().str.contains(last_name.lower(), regex=False))] 
    
    mentions_count = len(name_df)
    positive_count = len(name_df[name_df['Sentiment'] == 'positive'])
    neutral_count = len(name_df[name_df['Sentiment'] == 'neutral'])
    negative_count = len(name_df[name_df['Sentiment'] == 'negative'])
    
    if mentions_count > 0:
        sentiment_score = positive_count - negative_count
    else:
        sentiment_score = 0
        
    result_dict['player'].append(name)
    result_dict['position'].append(df_nameslist['position'].iloc[iteration])
    result_dict['team'].append(df_nameslist['team'].iloc[iteration])
    result_dict['age'].append(df_nameslist['age'].iloc[iteration])
    result_dict['mentions_count'].append(mentions_count)
    result_dict['sentiment_score'].append(sentiment_score)
    result_dict['positive_tweets'].append(positive_count)
    result_dict['neutral_tweets'].append(neutral_count)
    result_dict['negative_tweets'].append(negative_count)
    
    iteration += 1
```

```python
result_df = pd.DataFrame(result_dict)
df_sentiment_mentions = result_df
df_sentiment_mentions.to_csv('/Data/FIFA World Cup 2022 Twitter Dataset/sentiment_mentions.csv', index=False)
```
